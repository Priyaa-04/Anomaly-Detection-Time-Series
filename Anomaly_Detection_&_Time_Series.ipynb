1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.
     ### ***Anomaly Detection***

Anomaly Detection is a technique used to identify data points, events, or observations that deviate significantly from the normal pattern of a dataset. These unusual patterns are called ***outliers*** or ***anomalies***.

It is widely used in:

* Fraud detection
* Network intrusion detection
* Medical diagnosis
* Fault detection in machines
* Banking and cybersecurity


### *Types of Anomalies*

### *1. Point Anomaly (Global Outlier)*

A single data point that is significantly different from the rest of the dataset.

*Example:*

* In a dataset of daily temperatures (25°C–35°C), one value is 60°C.
* A credit card transaction of ₹5,00,000 when the customer usually spends ₹5,000–₹10,000.

### *2. Contextual Anomaly (Conditional Outlier)*

A data point that is abnormal in a specific context but may be normal in another context.

*Context = Time, Location, Season, etc.*

*Example:*

* 30°C temperature in summer → Normal
* 30°C temperature in winter → Anomaly
* High website traffic at 2 PM → Normal
* Same traffic at 2 AM → Suspicious

So, the anomaly depends on the *context*.


### *3. Collective Anomaly*

A group of related data points that are abnormal together, even if individual points may appear normal.

*Example:*

* A sudden spike in network traffic over 10 minutes indicating a DDoS attack.
* A sequence of small transactions that together indicate fraud.

2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.
     ### *1. Isolation Forest*

***Approach:***

* Based on random trees (ensemble method).
* Works by randomly selecting features and splitting data.
* Anomalies are isolated quickly (require fewer splits).

*Key Idea:*
Anomalies are rare and easier to isolate.

***Advantages:***

* Works well with high-dimensional data
* Fast and scalable
* No need for distance calculations

***Suitable Use Cases:***

* Fraud detection
* Large financial datasets
* High-dimensional cybersecurity data
### *2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)*

***Approach:***

* Density-based clustering algorithm
* Groups points that are close together
* Points in low-density regions are labeled as outliers

*Parameters:*

* eps (radius)
* min_samples (minimum points to form cluster)

***Advantages:***

* Can detect arbitrary shaped clusters
* Automatically identifies noise

***Limitations:***

* Sensitive to parameter selection
* Not ideal for varying densities

***Suitable Use Cases:***

* Spatial data
* GPS location clustering
* Image segmentation


### *3. Local Outlier Factor (LOF)*

***Approach:***

* Density-based method
* Compares local density of a point with its neighbors
* If density is much lower than neighbors → anomaly

*Key Idea:*
Detects *local anomalies*, not just global ones.

***Advantages:***

* Good for datasets with varying densities
* Detects subtle anomalies

***Limitations:***
* Computationally expensive
* Not ideal for very large datasets

***Suitable Use Cases:***

* Fraud detection
* Medical anomaly detection
* Detecting local deviations in data

3. What are the key components of a Time Series? Explain each with one
example.

    - A ***Time Series*** is a sequence of data points recorded at regular time intervals (daily, monthly, yearly, etc.).

A time series typically consists of ***four key components***:


### *1. Trend (T)*

The long-term movement or direction in the data over time.
It shows whether the data is increasing, decreasing, or stable.

*Example:*

* A company’s annual sales increasing every year from 2015 to 2025.
* Population growth over decades.

**Trend shows the overall upward or downward direction.**


### *2. Seasonality (S)*

Regular and repeating patterns that occur within a fixed time period (daily, monthly, yearly).

*Example:*

* Ice cream sales increase every summer.
* Retail sales increase every Diwali or Christmas season.

**Seasonal patterns repeat at fixed intervals.**


### *3. Cyclical Component (C)*

Long-term fluctuations that occur due to economic or business cycles.
These cycles do not have a fixed time period.

*Example:*

* Economic boom and recession cycles.
* Stock market rising and falling over several years.

**Cycles are irregular in duration but occur over long periods.**


### *4. Irregular (Random/Noise) Component (R)*

Unpredictable, random variations in the data caused by unexpected events.

*Example:*

* Sudden drop in sales due to COVID-19 lockdown.
* Natural disasters affecting production.

These are unpredictable and do not follow a pattern.

## *Time Series Model Representation*

Time series can be represented as:

*Additive Model:*
[
Y = T + S + C + R
]

*Multiplicative Model:*
[
Y = T \times S \times C \times R
]

* Use *Additive* when seasonal variation is constant.
* Use *Multiplicative* when seasonal variation increases with trend.


4. Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?
     ## ***Stationarity in Time Series***

A time series is said to be *stationary* if its statistical properties do not change over time.

This means:

1. Mean is constant
2. Variance is constant
3. Covariance depends only on lag, not time

Most forecasting models like ARIMA require stationary data.


## *Types of Non-Stationarity*

1. Trend non-stationarity
2. Seasonal non-stationarity
3. Changing variance


## *How to Test Stationarity*

### *1. Visual Inspection*

* Plot the time series.
* If mean or variance changes over time → Non-stationary.

### *2. Augmented Dickey-Fuller (ADF) Test*

* Null Hypothesis (H₀): Series is non-stationary.
* If p-value < 0.05 → Reject H₀ → Series is stationary.
* If p-value > 0.05 → Non-stationary.

### *3. KPSS Test*

* Null Hypothesis: Series is stationary.
* If p-value < 0.05 → Non-stationary.


## *How to Convert Non-Stationary Series to Stationary*

### *1. Differencing*

Subtract previous value from current value:  
[
Y_t' = Y_t - Y_{t-1}
]

Used to remove trend.


### *2. Log Transformation*

Used when variance increases over time:

[
Y' = \log(Y)
]

Reduces heteroscedasticity.

### *3. Seasonal Differencing*

Used to remove seasonality:

[
Y_t' = Y_t - Y_{t-m}
]

(where m = seasonal period)


### *4. Moving Average Smoothing*

Reduces noise and smooths fluctuations.


5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.

     Dataset:

     ● NYC Taxi Fare Data

     ● AirPassengers Dataset

# 1. AR (AutoRegressive) Model

**### Structure**

The AR model assumes the current value depends on its previous values.

[
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + ... + \epsilon_t
]

* Parameter: *p* (number of lag observations)
* Notation: *AR(p)*
* Works only with *stationary data*

**### Application**

### NYC Taxi Fare Data

If we take daily average taxi fares and remove trend/seasonality, AR can model short-term dependency (e.g., today's fare influenced by yesterday's fare).

###  AirPassengers Dataset

Not suitable directly because the dataset has strong trend and seasonality.


# 2. MA (Moving Average) Model

**### Structure**

The MA model assumes the current value depends on past forecast errors.

[
Y_t = c + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + ... + \epsilon_t
]

* Parameter: *q* (error lag)
* Notation: *MA(q)*
* Works on stationary series

**### Application**

###  NYC Taxi Fare Data

Useful when fares fluctuate due to temporary shocks (traffic congestion, bad weather, events).

###  AirPassengers Dataset

Not suitable alone due to non-stationarity and seasonality.


# 3. ARIMA (AutoRegressive Integrated Moving Average)

**### Structure**

ARIMA(p, d, q)

* p = AR order
* d = Differencing order (to make series stationary)
* q = MA order

ARIMA combines AR + differencing + MA.

**### Application**

### AirPassengers Dataset

* Shows strong upward trend.
* Needs differencing (d = 1).
* ARIMA can handle trend but does *not explicitly model seasonality*.

###NYC Taxi Fare Data

If taxi fares show long-term trend (e.g., increasing fares over years), ARIMA can model it after differencing.


# 4. SARIMA (Seasonal ARIMA)

**### Structure**

SARIMA(p, d, q)(P, D, Q, s)

* (p, d, q) → Non-seasonal part
* (P, D, Q, s) → Seasonal part
* s = seasonal period (e.g., 12 for monthly data)

**### Application**

### AirPassengers Dataset

* Monthly airline passenger data
* Clear yearly seasonality (s = 12)
* Trend + seasonal pattern present

✔ ***Best model for AirPassengers dataset***

### NYC Taxi Fare Data

If fares show weekly or monthly seasonal patterns (e.g., higher on weekends), SARIMA can model those patterns.

# 5. SARIMAX (Seasonal ARIMA with Exogenous Variables)

**### Structure**

SARIMAX = SARIMA + External Variables (X)

[
Y_t = ARIMA + Seasonal + \beta X_t
]

* Includes external predictors (exogenous variables)

**### Application**

###NYC Taxi Fare Data

Taxi fares depend on:

* Weather conditions
* Holidays
* Traffic
* Time of day

SARIMAX allows including these external variables, making predictions more accurate.

✔ ***Best model for NYC Taxi Fare Data***

###AirPassengers Dataset

If external factors (e.g., fuel prices, GDP, tourism index) are included, SARIMAX can be applied.


6.Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components

```python

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Rename column
df.columns = ['Passengers']

# Plot original series
plt.figure(figsize=(10,5))
plt.plot(df['Passengers'])
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# Decompose time series (multiplicative because variance increases)
decomposition = seasonal_decompose(df['Passengers'], model='multiplicative')

# Plot decomposition
decomposition.plot()
plt.show()
```

7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Create synthetic dataset (like taxi fare vs distance)
np.random.seed(42)
distance = np.random.normal(5, 2, 300)
fare = distance * 10 + np.random.normal(0, 5, 300)

# Add anomalies
distance[:10] = np.random.uniform(20, 30, 10)
fare[:10] = np.random.uniform(200, 300, 10)

data = pd.DataFrame({'Distance': distance, 'Fare': fare})

# Apply Isolation Forest
model = IsolationForest(contamination=0.05, random_state=42)
data['Anomaly'] = model.fit_predict(data[['Distance', 'Fare']])

# Plot results
plt.figure(figsize=(8,6))
plt.scatter(data['Distance'], data['Fare'],
            c=data['Anomaly'], cmap='coolwarm')
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Distance")
plt.ylabel("Fare")
plt.show()
```

8. Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
df.columns = ['Passengers']

# Fit SARIMA model (1,1,1)(1,1,1,12)
model = SARIMAX(df['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

results = model.fit()

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Plot original data
plt.figure(figsize=(10,5))
plt.plot(df['Passengers'], label='Original')
plt.plot(forecast_values, label='Forecast', color='red')
plt.fill_between(confidence_intervals.index,
                 confidence_intervals.iloc[:, 0],
                 confidence_intervals.iloc[:, 1],
                 color='pink', alpha=0.3)
plt.legend()
plt.title("SARIMA Forecast (Next 12 Months)")
plt.show()
```

9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Create synthetic dataset
np.random.seed(42)
X = 0.3 * np.random.randn(300, 2)

# Add outliers
X[:15] = np.random.uniform(low=-4, high=4, size=(15, 2))

# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

# Separate normal and anomalies
normal = X[y_pred == 1]
anomaly = X[y_pred == -1]

# Plot results
plt.figure(figsize=(8,6))
plt.scatter(normal[:, 0], normal[:, 1])
plt.scatter(anomaly[:, 0], anomaly[:, 1])
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
